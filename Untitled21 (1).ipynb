### Determining the value of K i.e Optimum number of clusters

```python
#Store the dat values in x using iloc
x = iris_df.iloc[:, [0, 1, 2, 3]].values
```

```python
from sklearn.cluster import KMeans
```

```python
#finding the value of K by plotting sum of squares in respective clusters wrt no of clusters
sum_of_squares = []
for i in range(1,11) :
    
    k_means = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    k_means.fit(x)
    sum_of_squares.append(k_means.inertia_)
# Elbow method to determine the value of k   
plt.plot(range(1,11),sum_of_squares)
plt.title("Elbow Method to determine K")
plt.ylabel("sum_of_squares")
plt.xlabel("No. of clusters")
plt.show()
```

## Plotting the clusters

```python
# Using elbow method, K=3
k_means = KMeans(n_clusters = 3, init = 'k-means++',
                max_iter = 300, n_init = 10, random_state = 0)
```

```python
#using fit_predict to store the resspective cluster no. for each value
y_kmeans = k_means.fit_predict(x)
```

```python
#plotting the different clusters (between sepal length and sepal width)
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], 
            s = 100, c = 'tomato', label = 'Iris-setosa')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], 
            s = 100, c = 'cadetblue', label = 'Iris-versicolour')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1],
            s = 100, c = 'limegreen', label = 'Iris-virginica')

# Plotting the centroids of the clusters
plt.scatter(k_means.cluster_centers_[:, 0], k_means.cluster_centers_[:,1], 
            s = 100, c = 'yellow', label = 'Centroids')

plt.legend()
```

```python
#plotting the different clusters (between sepal length and petal length)
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 2], 
            s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 2], 
            s = 100, c = 'blue', label = 'Iris-versicolour')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 2],
            s = 100, c = 'green', label = 'Iris-virginica')

# Plotting the centroids of the clusters
plt.scatter(k_means.cluster_centers_[:, 0], k_means.cluster_centers_[:,2], 
            s = 100, c = 'yellow', label = 'Centroids')

plt.legend()
```

```python
#plotting the different clusters (between sepal width and petal width)
plt.scatter(x[y_kmeans == 0, 1], x[y_kmeans == 0, 3], 
            s = 100, c = 'gray', label = 'Iris-setosa')
plt.scatter(x[y_kmeans == 1, 1], x[y_kmeans == 1, 3], 
            s = 100, c = 'hotpink', label = 'Iris-versicolour')
plt.scatter(x[y_kmeans == 2, 1] ,x[y_kmeans == 2, 3], s = 100, c = 'aqua', label = 'Iris-virginica')

# Plotting the centroids of the clusters
plt.scatter(k_means.cluster_centers_[:, 1], k_means.cluster_centers_[:,3], 
            s = 100, c = 'yellow', label = 'Centroids')

plt.legend()
```
